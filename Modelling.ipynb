**Problem:** Customer churn prediction (target column: `is_churn`, where `1` indicates churn and `0` indicates non-churn)

**Dataset:** `final_data.csv`

```python
# Load & info dataset

import pandas as pd
import matplotlib.pyplot as plt
pd.set_option("display.max_columns", None)
df = pd.read_csv("/content/final_data.csv")
display(df.head())

df.info()

duplicates = df.duplicated().sum()
print(f"Duplicate rows: {duplicates}")

display(df.describe())

duplicates = df.duplicated().sum()
print(f"Duplicate rows: {duplicates}")

df["is_churn"].value_counts().plot(kind="bar")
plt.title("Target Distribution")
plt.xlabel("is_churn")
plt.ylabel("Count")
plt.show()
print(df["is_churn"].value_counts(normalize=True)*100)
```

```python
# Replace invalid gender value
df["gender"] = df["gender"].replace([0, "0"], "unknown")

# Remove invalid records
df = df[df["registration_duration"] >= 0].copy()

# Drop identifier and original datetime columns
df = df.drop(columns=[
    "msno",
    "registration_init_time",
    "last_membership_expire",
    "last_user_log_date"
])

# Convert categorical features to string
df["city"] = df["city"].astype(int).astype(str)
df["registered_via"] = df["registered_via"].astype(int).astype(str)
df["gender"] = df["gender"].astype(str)
```

```python
# Split train test
from sklearn.model_selection import train_test_split
X = df.drop(columns="is_churn")
y = df["is_churn"]
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
print(f"Training set : {X_train.shape}")
print(f"Testing set  : {X_test.shape}")
```

```python
# Scaling & encoding
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
cat_cols = ["gender", "city", "registered_via"]
num_cols = [col for col in X_train.columns if col not in cat_cols]
# Logistic Regression
preprocessor_logreg = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore", drop="first"), cat_cols)
    ]
)
# XGBoost
preprocessor_xgb = ColumnTransformer(
    transformers=[
        ("num", "passthrough", num_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols)
    ]
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
neg = (y_train == 0).sum()
pos = (y_train == 1).sum()
scale_pos_weight = neg / pos
# Logistic Regression
logreg = Pipeline([
    ("preprocessor", preprocessor_logreg),
    ("classifier", LogisticRegression(
        class_weight="balanced",
        max_iter=1000,
        random_state=42
    ))
])
# XGBoost
xgb = Pipeline([
    ("preprocessor", preprocessor_xgb),
    ("classifier", XGBClassifier(
        n_estimators=300,
        learning_rate=0.05,
        max_depth=6,
        subsample=0.8,
        colsample_bytree=0.8,
        scale_pos_weight=scale_pos_weight,
        eval_metric="logloss",
        random_state=42
    ))
])
logreg.fit(X_train, y_train)
xgb.fit(X_train, y_train)
```

```python
# Model Evaluation
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    classification_report,
    confusion_matrix
)
def evaluate_model(model, X_test, y_test):
    pred = model.predict(X_test)
    prob = model.predict_proba(X_test)[:,1]
    print(classification_report(y_test, pred))
    print(confusion_matrix(y_test, pred))
    return {
        "Accuracy": accuracy_score(y_test, pred),
        "Precision": precision_score(y_test, pred),
        "Recall": recall_score(y_test, pred),
        "F1-score": f1_score(y_test, pred),
        "ROC AUC": roc_auc_score(y_test, prob)
    }
result_lr = evaluate_model(logreg, X_test, y_test)
result_xgb = evaluate_model(xgb, X_test, y_test)
```

```python
# Model comparison
results = pd.DataFrame(
    [result_lr, result_xgb],
    index=["Logistic Regression", "XGBoost"]
)
display(results.round(4))
```
