```python
import numpy as np

transactions["payment_plan_days"].value_counts().sort_index()
transactions.loc[transactions["payment_plan_days"] <= 0, "payment_plan_days"].value_counts()
transactions.loc[transactions["payment_plan_days"] <= 0, "payment_plan_days"] = np.nan
```

```python
invalid_date = transactions[
    transactions["transaction_date"] > transactions["membership_expire_date"]
]
len(invalid_date)

transactions = transactions[
    transactions["transaction_date"] <= transactions["membership_expire_date"]
]
transactions['membership_expire_date'] = pd.to_datetime(
    transactions['membership_expire_date'].astype(str),
    format='%Y%m%d'
)
```

```python
transactions['transaction_date'] = pd.to_datetime(
    transactions['transaction_date'].astype(str),
    format='%Y%m%d'
)

cutoff_date = pd.to_datetime("2017-04-01")
transactions = transactions[
    transactions['transaction_date'] < cutoff_date
]
```

```python
userlogs['date'] = pd.to_datetime(
    userlogs['date'].astype(str),
    format='%Y%m%d'
)

userlogs = userlogs[userlogs['date'] < cutoff_date]
```

Feature Engineering

```python
# Pastikan transaction_date & membership_expire_date sudah datetime
transactions['transaction_date'] = pd.to_datetime(transactions['transaction_date'], format='%Y%m%d')
transactions['membership_expire_date'] = pd.to_datetime(transactions['membership_expire_date'], format='%Y%m%d')

# Agregasi fitur transaksi
trans_agg = transactions.groupby('msno').agg(
    total_transactions=('transaction_date', 'count'),
    total_payment=('actual_amount_paid', 'sum'),
    avg_plan_days=('payment_plan_days', 'mean'),
    cancel_rate=('is_cancel', 'mean'),
    auto_renew_rate=('is_auto_renew', 'mean'),
    last_membership_expire=('membership_expire_date', 'max'),
    avg_amt_per_day=('actual_amount_paid', lambda x: (x / transactions.loc[x.index, 'payment_plan_days']).mean())
   ).reset_index()

# Indicator churn / loyal
last_trans = transactions.sort_values('transaction_date').groupby('msno').last().reset_index()
# Ambil transaksi TERAKHIR per user
last_trans = (
    transactions
    .sort_values('transaction_date')
    .groupby('msno')
    .last()
    .reset_index()
)

last_trans['autorenew_and_not_cancel'] = (
    (last_trans['is_auto_renew'] == 1) &
    (last_trans['is_cancel'] == 0)
).astype(int)

# MERGE ke trans_agg
trans_agg = trans_agg.merge(
    last_trans[['msno', 'autorenew_and_not_cancel']],
    on='msno',
    how='left'
)

membership_duration = (
    transactions.assign(
        plan_days=(transactions['membership_expire_date'] - transactions['transaction_date']).dt.days
    )
    .groupby('msno')['plan_days']
    .mean()
    .reset_index(name='membership_duration')
)

trans_agg = trans_agg.merge(membership_duration, on='msno', how='left')

# Long time user (lebih dari 1 tahun sejak registrasi)
members['registration_init_time'] = pd.to_datetime(members['registration_init_time'], format='%Y%m%d')
members['registration_duration'] = (cutoff_date - members['registration_init_time']).dt.days
members['long_time_user'] = (members['registration_duration'] > 365).astype(int)

# Jarak hari sejak transaksi terakhir
trans_agg['days_since_last_transaction'] = (cutoff_date - last_trans['transaction_date']).dt.days
```

```python
# Tambah kolom diskon per transaksi
transactions['discount'] = transactions['plan_list_price'] - transactions['actual_amount_paid']
transactions['is_discount'] = (transactions['discount'] > 0).astype(int)

# Agregasi persenan transaksi dapat diskon per user
discount_pct = transactions.groupby('msno')['is_discount'].mean().reset_index()
discount_pct.rename(columns={'is_discount': 'pct_discount_transactions'}, inplace=True)

# Merge ke trans_agg
trans_agg = trans_agg.merge(discount_pct, on='msno', how='left')
```

```python
userlog_agg = userlogs.groupby('msno').agg(
    total_listening_secs=('total_secs', 'sum'),
    avg_listening_secs=('total_secs', 'mean'),
    active_days=('date', 'nunique'),
    last_user_log_date=('date', 'max')  # tanggal terakhir login
).reset_index()
```

```python
final_data = (
    train
    .merge(members, on='msno', how='left')
    .merge(trans_agg, on='msno', how='left')
    .merge(userlog_agg, on='msno', how='left')
)

final_data.fillna(0, inplace=True)
final_data.shape
```
